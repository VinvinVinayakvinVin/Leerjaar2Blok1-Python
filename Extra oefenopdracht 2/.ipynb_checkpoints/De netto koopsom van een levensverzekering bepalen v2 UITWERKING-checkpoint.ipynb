```python
import pandas as pd
import math
import matplotlib.pyplot as plt

# Zelf geschreven functies die gebruikt worden:

# Bepalen welke rijindex de leeftijd van de deelnemer heeft
def frowIndex(lftDln):
    for i in range(len(progMan_tabel)):
        if progMan_lijst[i][0] == lftDln:
            return i

# Lijst maken met erin Qx, Px en nPx
def kansenBasis(geslacht, aantalRijen, rowIndex):
    if geslacht == 'v':
            lijst = progVrouw_lijst
    else:
        lijst = progMan_lijst
    kansen = []
    for i in range(aantalRijen):
        rij = []
        Qx = lijst[rowIndex + i][kolomIndex + i]
        rij.append(Qx)
        rij.append(1 - Qx)
        if i == 0:
            rij.append(1)
        else:
            rij.append(kansen[i - 1][1] * kansen[i - 1][2])
        kansen.append(rij)  
    return kansen

# Lijst maken met erin nPxy en kans één leeft én één dood
def kansenSamen(kansenDlnm1, kansenDlnm2):
    kansen = []
    for i in range(maxaantalRijen):
        rij = []
        # leeftijd + i <= 120
        if i < minaantalRijen:
            nPxy = kansenDlnm1[i][2] * kansenDlnm2[i][2]
        # leeftijd + i > 120 voor een deelnemer, daar staan geen kansen meer in. Iemand wordt nooit 121.
        else:
            nPxy = 0 
        rij.append(nPxy)
        # leeftijd + i <= 120
        if i < minaantalRijen:
                    nPlft_dood = kansenDlnm1[i][2] * (1 - kansenDlnm2[i][2]) + kansenDlnm2[i][2] * (1- kansenDlnm1[i][2])
        # leeftijd + i > 120 voor een deelnemer, daar staan geen kansen meer in. Dus alleen op leven van de levende nog kijken.
        else:
            if aantalRijenDlnm1 == minaantalRijen:
                nPlft_dood = kansenDlnm2[i][2]
            else:
                nPlft_dood = kansenDlnm1[i][2]
        rij.append(nPlft_dood)
        kansen.append(rij)
    return kansen
    
# Contant maken met rts:
def contantMaken(kansen):
    som = 0
    i = 0
    for kans in kansen:
        rente = rts_lijst[i][1]
        rentecontant = 1 / math.pow((1 + rente), i)
        i += 1
        jaarcontant = ((uitk_Beideleven * kans[0] + uitk_Eenleeft * kans[1])) * rentecontant
        som = som + jaarcontant
    return som

# Excelsheets inlezen: prognosetafel man en vrouw en RTS
rts_tabel = pd.read_excel('Prognosetafel en rts.xlsx', sheet_name = 'RTS')
progMan_tabel = pd.read_excel('Prognosetafel en rts.xlsx', sheet_name = 'Progman')
progVrouw_tabel = pd.read_excel('Prognosetafel en rts.xlsx', sheet_name = 'Progvrouw')

# Voor als je wil kijken hoe het dataframe eruit ziet van de ingelezen excel-sheets:
# display(rts_tabel)
# display(progMan_tabel)
# display(progVrouw)

# Van de dataframes lijsten maken, let op de kolomnamen komen er niet in te staan, dus
# vandaar ook de .colums gebruikt om een lijst tge krijgen met de jaartallen
rts_lijst = rts_tabel.values.tolist()
Kolom_lijst = progMan_tabel.columns.values.tolist()
progMan_lijst = progMan_tabel.values.tolist()
progVrouw_lijst = progVrouw_tabel.values.tolist()

# Voor als je wil kijken hoe de lists eruit zien van de dataframes van de excel-sheets:
# display(rts_lijst)
# display(progMan_lijst)
# display(progVrouw_lijst)

# Persoonlijke gegevens inlezen:
lftDln1 = int(input("Vul de leeftijd in van deelnemer 1: "))
geslDln1 = input("Vul het geslacht (m/v) in van deelnemer 1: ")
lftDln2 = int(input("Vul de leeftijd in van deelnemer 2: "))
geslDln2 = input("Vul het geslacht (m/v) in van deelnemer 2: ")

# Overige gegevens:
rekenjaar = int(input("Vul het rekenjaar in: "))
uitk_Beideleven = int(input("Vul de uitkering in als beide leven: "))
uitk_Eenleeft = int(input("Vul de uitkering in als slechts één leeft: "))

# Lijst maken voor dln1 met erin Qx, Px en nPx, daarna zelfde lijst maar dan vooir dln2:
rowIndexDlnm1 = frowIndex(lftDln1)
aantalRijenDlnm1 = len(progMan_tabel) - rowIndexDlnm1
rowIndexDlnm2 = frowIndex(lftDln2)
aantalRijenDlnm2 = len(progMan_tabel) - rowIndexDlnm2
kolomIndex = Kolom_lijst.index(rekenjaar)
minaantalRijen = min(aantalRijenDlnm1, aantalRijenDlnm2)
maxaantalRijen = max(aantalRijenDlnm1, aantalRijenDlnm2)    

kansenDlnm1 = kansenBasis(geslDln1, aantalRijenDlnm1, rowIndexDlnm1)
kansenDlnm2 = kansenBasis(geslDln2, aantalRijenDlnm2, rowIndexDlnm2)

# Lijst maken voor nPxy en de kans dat er één leeft en één dood is:    
kansen = kansenSamen(kansenDlnm1, kansenDlnm2)

# Contante waarde en afgerond op twee decimalen, daarna uitprinten.
contanteWaarde = round(contantMaken(kansen), 2) 

print(f"De koopsom voor de verzekering is: {contanteWaarde:,}")

# zonder duizendseparator:
# print('De koopsom voor de verzekering is:', round(contantMaken(kansen), 2))

# Grafieken van sterftekansen maken
QxDlnm1 = []
PxDlnm1 = []
leeftijdDlnm1 = []
QxDlnm2 = []
PxDlnm2 = []
leeftijdDlnm2 = []

i = 0
for rij in kansenDlnm1:
    QxDlnm1.append(rij[0])
    PxDlnm1.append(rij[1])
    leeftijdDlnm1.append(lftDln1 + i)
    i += 1

i = 0
for rij in kansenDlnm2:
    QxDlnm2.append(rij[0])
    PxDlnm2.append(rij[1]) 
    leeftijdDlnm2.append(lftDln2 + i)
    i += 1
    
plt.plot(leeftijdDlnm1, QxDlnm1, label = 'Qx deelnemer 1', color = 'lightskyblue')
plt.plot(leeftijdDlnm1, PxDlnm1, label = 'Px deelnemer 1', color = 'darkblue') # kleur opzoeken op internet
plt.plot(leeftijdDlnm2, QxDlnm2, label = 'Qx deelnemer 2', color = 'pink')
plt.plot(leeftijdDlnm2, PxDlnm2, label = 'Px deelnemer 2', color = 'hotpink')

plt.xlabel('Leeftijd')
plt.ylabel('Kans')

plt.legend()
```
